# 2. Random Variables
Probability is a language that lets you communicate information about uncertain outcomes and events. By assigning a numeric value between zero and one to an event, or a collection of outcomes, probability measures how likely an event is to occur. Probability is immensely useful  and adheres to only a few basic rules.  Assigning probabilities to events, however, is a challenging problem.  First known applications of probability are in games that involve a chance, e.g. dice roll or card draw. 

Random variables are quantities that we are not certain about. The simplest version of a random variable is a binary yes/no outcome. Another example of a random variable is temperature tomorrow. The temperature variable can take a continuum of values and is called a continuous random variable. 

A random variable then will describe an uncertain quantity by attaching a numeric value to the occurrence of an event. Two examples of binary random variable are
- Will a user click-through on a Google ad? 
- Will Trump win 2020 elections?

The quintessential random event is an outcome of a coin toss. Possible outcomes are Head (H) and Tails (T), each having a probability of 1/2.  

However, there is nothing random about a coin flip. You can see Persi Diaconis flipping a coin and getting heads every single time. Before becoming a probability professor, he was a magician!  [Video](https://youtu.be/ZdA7TeoZD_g?t=106).

There are three sources of randomness
- Our inability to control the process that generates an outcome, e.g. coin flip
- Our inability to measure accurately the outcome, e.g. we cannot ask every person in the work which design he or she prefers
- Truly random phenomenon, which is very rare!

Now the question is, do we ever have objective uncertainty, or a random variable always a deterministic quantity inaccurately measured? Two examples of truly random phenomenas include quantum mechanics, nuclear decay. Nuclear decay is impossible to predict and is impossible to predict according to our current physical knowledge of this process. In fact, the outcomes of nuclear decay process are used by on-line poker rooms as the most reliable source of random numbers. Some also can argue that human decision making is a random process. A relatively new interdisciplinary field called neuroeconomics studies the question of randomness of our decisions.

In AB testing and in statistics in general we are often interested in averages. For example, we might be interested in how people in Michigan will vote during next presidential elections. We can ask every person in Michigan and record their answers as a sequence of 0 (Dem.) and 1 (Rep.), assume there are no undecided voters. In 2016 Clinton got 49.9% and Trump got 50.1%. What will happen in 2020?

One of the foundational ideas in AB testing and in statistics in general is that we can make conclusions about Michigan vote by asking a small number of people (sample). Now we introduce the concept of sampling and discuss how to deal with uncertainty associated with sampling. 

We will use simulations to demonstrate the concept. Say we went to Michigan and asked 100 people about their voting preferences for 2020. 

```r
x = rbinom(n = 100, size = 1, prob = 0.501)
mean(x)
```

Say next month we go and ask different 100 people

```r
x = rbinom(n = 100, size = 1, prob = 0.501)
mean(x)
```

- The resulrs for two groups of 100 people are very different! 
- This is called **sampling error**. 
- If we can ask every person in Michigan we will not have any error but it is impractical. 
- Nate Silver made a career by adjusting those sampling errors. 


What happens if we increase the size of the group of people we ask to 10000?

```r
x = rbinom(n = 10000, size = 1, prob = 0.501)
mean(x)
```

```r
x = rbinom(n = 10000, size = 1, prob = 0.501)
mean(x)
```

The results from two groups of 10000 are much closer! The sampling error can be calculated and it is proportional to $1/\sqrt{n}$. Where $n$ is the sumber of people in the sample. Let's see how the histogram of those errors look like. Say we ask different group of people 5000 times. Computer simulations allow us to do that :)

If you want to double accuracy, you have to quadruple sample size.

```r
means = rep(0,5000)
for (i in 1:5000) means[i] = mean(rbinom(n = 100, size = 1, prob = 0.501))
hist(means, main="", breaks=40)
```

```r
means = rep(0,5000)
for (i in 1:5000) means[i] = mean(rbinom(n = 10000, size = 1, prob = 0.501))
hist(means, main="", breaks=40)
```

A few observations:
- Sampling error is smaller when sample size is bigger, dah!
- Sampling errors are equally likely to be positive and negative (histograms are symmetric)
- Histogram has what in statistics is called a Bell-shape or more scientifically Normal distribution
  - The rate at which errors decay is exponential.

Tools of statistics allow us to quantify the uncertainty about the average vote and calculate what is called a confidence interval. Say we collected a sample $x_1,x_2,\ldots,x_n$. Then the average is 

$$
\bar{x} = \frac{x_1 + x_2 + \ldots + x_n}{n}
$$

The formula for confidence interval for $\bar{x}$ is rather simple
$$
\bar{x} \pm 1.96\times SEM
$$

According to the central limit theorem, if we are to repeat our survey many, many times, in <font color=green>95% of the cases our estimated mean will be inside this confidence interval!</font>
$$
[\bar{x} - 1.96\times SEM, \bar{x} + 1.96\times SEM]
$$

Here $SEM$ is called standard error of the mean and can be calculated as 
$$
SEM = \frac{\sigma}{\sqrt{n}}
$$
And $\sigma$ is the standard deviation computed from the sample
$$
\sigma^2 = \frac{(x_1 - \bar{x})^2 + (x_2 - \bar{x})^2 + \ldots + (x_n - \bar{x})^2}{n}
$$

Let's apply those formulas!

```r
x = rbinom(n = 10000, size = 1, prob = 0.501)
xbar = sum(x)/10000
sigma2 = sum((x-xbar)^2)/10000
sigma = sqrt(sigma2)
SEM = sigma/sqrt(10000)

sprintf("xbar is %.2f", xbar)
sprintf("sigma is %.2f", sigma)
sprintf("SEM is %.2f", SEM)
sprintf("The confidence interval is [%.2f,%.2f]", xbar-1.96*SEM, xbar+1.96*SEM)
```

```r
x = rbinom(n = 1000, size = 1, prob = 0.501)
xbar = sum(x)/1000
sigma2 = sum((x-xbar)^2)/1000
sigma = sqrt(sigma2)
SEM = sigma/sqrt(1000)

sprintf("xbar is %.2f", xbar)
sprintf("sigma is %.2f", sigma)
sprintf("SEM is %.2f", SEM)
sprintf("The confidence interval is [%.2f,%.2f]", xbar-1.96*SEM, xbar+1.96*SEM)
```
